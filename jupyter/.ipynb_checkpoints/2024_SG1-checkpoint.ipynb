```python
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from astropy.coordinates import SkyCoord
import astropy.units as u

def fetch_horizons_data(start_time, stop_time, step_size='1d', des_id='54481740'):
    url = "https://ssd.jpl.nasa.gov/api/horizons.api"
    params = {
        'format': 'text',
        'COMMAND': f"'DES={des_id};'",
        'OBJ_DATA': 'YES',
        'MAKE_EPHEM': 'YES',
        'EPHEM_TYPE': 'OBSERVER',
        'CENTER': '500@399',  # Earth-centered observer
        'START_TIME': start_time,
        'STOP_TIME': stop_time,
        'STEP_SIZE': step_size,
        'QUANTITIES': '1'
    }
    response = requests.get(url, params=params)
    return response.text if response.status_code == 200 else None

def parse_horizons_data(data):
    lines = data.split('\n')
    start_idx = lines.index('$$SOE') + 1
    end_idx = lines.index('$$EOE')
    
    dates, ra, dec = [], [], []
    for line in lines[start_idx:end_idx]:
        parts = line.split()
        dates.append(parts[0])
        ra.append(' '.join(parts[2:5]))
        dec.append(' '.join(parts[5:8]))
    
    return dates, ra, dec

def plot_trajectory(dates, ra, dec, title="Asteroid Trajectory"):
    coords = SkyCoord(ra=ra, dec=dec, unit=(u.hourangle, u.deg))
    x = coords.cartesian.x.value
    y = coords.cartesian.y.value
    z = coords.cartesian.z.value

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z)
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(title)
    
    plt.show()

# Fetch data only for spkid: 54481740
des_id = '54481740'
data = fetch_horizons_data('2006-01-01', '2007-01-20', des_id=des_id)

if data:
    try:
        dates, ra, dec = parse_horizons_data(data)
        plot_trajectory(dates, ra, dec, title=f"Asteroid {des_id} Trajectory")
        print(f"Data fetched and plotted successfully for ID: {des_id}")
    except Exception as e:
        print(f"Error parsing data for ID: {des_id} - {e}")
else:
    print(f"Failed to fetch data for ID: {des_id}")
```

```python
import requests
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from astropy.coordinates import SkyCoord
import astropy.units as u

def fetch_horizons_data(start_time, stop_time, step_size='1d', des_id='54481740'):
    url = "https://ssd.jpl.nasa.gov/api/horizons.api"
    params = {
        'format': 'text',
        'COMMAND': f"'DES={des_id};'",
        'OBJ_DATA': 'YES',
        'MAKE_EPHEM': 'YES',
        'EPHEM_TYPE': 'OBSERVER',
        'CENTER': '500@399',  # Earth-centered observer
        'START_TIME': start_time,
        'STOP_TIME': stop_time,
        'STEP_SIZE': step_size,
        'QUANTITIES': '1'
    }
    response = requests.get(url, params=params)
    return response.text if response.status_code == 200 else None

def parse_horizons_data(data):
    lines = data.split('\n')
    start_idx = lines.index('$$SOE') + 1
    end_idx = lines.index('$$EOE')
    
    dates, ra, dec = [], [], []
    for line in lines[start_idx:end_idx]:
        parts = line.split()
        dates.append(parts[0])
        ra.append(' '.join(parts[2:5]))
        dec.append(' '.join(parts[5:8]))
    
    return dates, ra, dec

def plot_trajectory_3d(dates, ra, dec, title="Asteroid Trajectory"):
    coords = SkyCoord(ra=ra, dec=dec, unit=(u.hourangle, u.deg))
    x = coords.cartesian.x.value
    y = coords.cartesian.y.value
    z = coords.cartesian.z.value

    # Create the trajectory line
    trajectory = go.Scatter3d(
        x=x, y=y, z=z,
        mode='lines',
        line=dict(color=z, colorscale='Viridis'),
        name='Asteroid Path'
    )

    # Add Sun at the origin
    sun = go.Scatter3d(
        x=[0], y=[0], z=[0],
        mode='markers',
        marker=dict(size=10, color='yellow'),
        name='Sun'
    )

    # Add Earth (approximate position)
    earth = go.Scatter3d(
        x=[1], y=[0], z=[0],  # 1 AU from Sun
        mode='markers',
        marker=dict(size=5, color='blue'),
        name='Earth'
    )

    # Create the layout
    layout = go.Layout(
        title=title,
        scene=dict(
            xaxis_title='X (AU)',
            yaxis_title='Y (AU)',
            zaxis_title='Z (AU)',
            aspectmode='data'
        ),
        height=800,
    )

    # Create the figure and add the traces
    fig = go.Figure(data=[trajectory, sun, earth], layout=layout)

    # Show the plot
    fig.show()

# Fetch data for spkid: 54481740
des_id = '54481740'
data = fetch_horizons_data('2006-01-01', '2007-01-20', des_id=des_id)

if data:
    try:
        dates, ra, dec = parse_horizons_data(data)
        plot_trajectory_3d(dates, ra, dec, title=f"Asteroid {des_id} Trajectory")
        print(f"Data fetched and plotted successfully for ID: {des_id}")
    except Exception as e:
        print(f"Error parsing data for ID: {des_id} - {e}")
else:
    print(f"Failed to fetch data for ID: {des_id}")
```
