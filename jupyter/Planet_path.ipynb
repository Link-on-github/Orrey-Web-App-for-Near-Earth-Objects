```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the orbit
a = 5  # Semi-major axis
e = 0.6  # Eccentricity

# Calculate semi-minor axis
b = a * np.sqrt(1 - e**2)

# Time parameter t (ranges from 0 to 2*pi for one orbit)
t = np.linspace(0, 2 * np.pi, 1000)

# Parametric equations for the elliptical orbit
x = a * np.cos(t)
y = b * np.sin(t)

# Plot the elliptical orbit
plt.plot(x, y, label="Planet's Orbit")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Elliptical Orbit of a Planet')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

planets = {
    "Mercury": {"a": 0.39, "e": 0.206},
    "Venus": {"a": 0.72, "e": 0.007},
    "Earth": {"a": 1.00, "e": 0.017},
    "Mars": {"a": 1.52, "e": 0.093},
    "Jupiter": {"a": 5.20, "e": 0.049},
    "Saturn": {"a": 9.58, "e": 0.056},
    "Uranus": {"a": 19.22, "e": 0.046},
    "Neptune": {"a": 30.05, "e": 0.010}
}

t = np.linspace(0, 2 * np.pi, 1000)

plt.figure(figsize=(8, 8))

plt.plot(0, 0, 'yo', markersize=12, label='Sun')

for planet, params in planets.items():
    a = params["a"]
    e = params["e"]
    b = a * np.sqrt(1 - e**2)
    
    x = a * np.cos(t)
    y = b * np.sin(t)
    
    plt.plot(x, y, label=planet)

plt.xlabel('x (AU)')
plt.ylabel('y (AU)')
plt.title('Orbit :)')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.legend()
plt.show()
```
